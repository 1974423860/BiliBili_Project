```python
# 综合评分和播放量对比
# 综合评分条形图
#p = plt.figure(figsize=(16,16))
fig,ax1 = plt.subplots()
plt.bar(anime_data['番剧名'],anime_data['评 分'],color='cyan')
plt.title('综合评分和播放量数据分析')
#ax1.tick_params(labelsize=6)

plt.xlabel('番剧名')
plt.ylabel('综合评分')
plt.xticks(fontsize=6)
plt.xticks(rotation=90,color='green') # 设置横坐标变量名旋转度数和颜色

# 播放量折线图
ax2 = ax1.twinx() # 组合图
ax2.plot(anime_data['总观看数'],color='red')
plt.ylabel('播放量')
plt.plot(1, label='综合评分', color="cyan", linewidth=5.0)
plt.plot(1, label='播放量', color="red", linewidth=1.0, linestyle="-")
plt.legend()
plt.savefig(r'./分析图/评分和播放量.png', dpi=1000, bbox_inches='tight')
plt.show()
```

```python
# 设置字体
pf = [0, 0, 0, 0, 0, 0, 0] #用来记录特定条件出现的次数
#开始逐个分析评分
for i in range(len(anime_data['评 分'])):
    if anime_data['评 分'][i] > 9.5:
        pf[0] = pf[0] + 1
    if anime_data['评 分'][i] > 9.0 and anime_data['评 分'][i] <= 9.5:
        pf[1] = pf[1] + 1
    if anime_data['评 分'][i] > 8.5 and anime_data['评 分'][i] <= 9.0:
        pf[2] = pf[2] + 1
    if anime_data['评 分'][i] > 8.0 and anime_data['评 分'][i] <= 8.5:
        pf[3] = pf[3] + 1
    if anime_data['评 分'][i] > 7.5 and anime_data['评 分'][i] <= 8.0:
        pf[4] = pf[4] + 1
    if anime_data['评 分'][i] > 7.0 and anime_data['评 分'][i] <= 7.5:
        pf[5] = pf[5] + 1
    if anime_data['评 分'][i] <= 7.0:
        pf[6] = pf[6] + 1

label = '>9.5', '9.0-9.5', '8.5-9.0', '8.0-8.5', '7.5-8.0', '7.0-7.5', '<7.0'#标题
color = 'red', 'orange', 'yellow', 'pink', 'green', 'blue'  #颜色
cs = [0, 0, 0, 0, 0, 0, 0]
#用来显示百分比占比
explode = [0, 0, 0, 0, 0, 0, 0]
#每块离中心的距离；
for i in range(0,7):  # 计算
    #print(i)
    cs[i] = pf[i] * 2
    explode[i] = pf[i] / 500

#开始配置图形
pie = plt.pie(cs, colors=color, explode=explode, labels=label, shadow=True, autopct='%1.1f%%')
for font in pie[1]:
    font.set_size(12)
for digit in pie[2]:
    digit.set_size(8)
plt.axis('equal')
#配置标题，字号大小等：
plt.title(u'各评分占比', fontsize=18)
plt.legend(loc=0, bbox_to_anchor=(0.8, 1))
leg = plt.gca().get_legend()
ltext = leg.get_texts()
plt.setp(ltext, fontsize=8)
#显示视图
plt.savefig('./分析图/各评分占比饼图.png')
plt.show()
```
